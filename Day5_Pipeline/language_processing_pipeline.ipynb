```python
import spacy
```

```python
nlp = spacy.blank("en")
doc = nlp("Captain America ate 100$ worth of samosas. Then, he said I can do this all day")

for token in doc:
  print(token)
```

```python
nlp.pipe_names
```

Although we have a blank pipeline, the tokenizer comes with it

```python
nlp = spacy.load("en_core_web_sm")
```

```python
nlp.pipe_names
```

```python
doc = nlp("Captain America ate 100$ worth of samosas. Then, he said I can do this all day")

for token in doc:
  print(token, " | ", token.pos_, " | ", token.lemma_)
```

```python
doc = nlp("Tesla Inc is going to acquire twitter for $30 billion")

for ent in doc.ents:
  print(ent.text, " | ", ent.label, " | ", spacy.explain(ent.label_))
```

```python
from spacy import displacy
displacy.render(doc, style='ent')
```
