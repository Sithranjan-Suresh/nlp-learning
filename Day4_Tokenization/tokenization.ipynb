```python
with open("students.txt") as f:
  text = f.readlines()
text
```

```python
text = ' '.join(text)
text
```

```python
doc = nlp(text)
emails = []
for token in doc:
  if token.like_email:
    print(token)
    emails.append(token.text)
```

```python
nlp = spacy.blank("hi")
doc = nlp("भैया जी! 5000 ₹ उधार थे वो वापस देदो")
for token in doc:
    print(token, token.is_currency, token.like_num)
```

```python
doc = nlp("gimme double cheese with extra large healthy pizza")
arr = [token.text for token in doc]
arr
```

```python
from spacy.symbols import ORTH
nlp.tokenizer.add_special_case("gimme", [
    {ORTH: "gim"},
    {ORTH: "me"}
])
```

```python
doc = nlp("gimme double cheese with extra large healthy pizza")
arr = [token.text for token in doc]
arr
```
