```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("Dr. Strange loves brisket of Texas. Hulk loves chicken of kentucky")
for sentence in doc.sents:
  print(sentence)
```

```python
for sentence in doc.sents:
  print(sentence)
  for word in sentence:
    print(word)
```
