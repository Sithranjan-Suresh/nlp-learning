```python
matches = re.findall(overall_pattern,chat2)
matches
```

```python
matches = re.findall(overall_pattern,chat3)
matches
```

```python
text='''
Born	Elon Reeve Musk
June 28, 1971 (age 50)
Pretoria, Transvaal, South Africa
Citizenship
South Africa (1971–present)
Canada (1971–present)
United States (2002–present)
Education	University of Pennsylvania (BS, BA)
Title
Founder, CEO and Chief Engineer of SpaceX
CEO and product architect of Tesla, Inc.
Founder of The Boring Company and X.com (now part of PayPal)
Co-founder of Neuralink, OpenAI, and Zip2
Spouse(s)
Justine Wilson
​
​(m. 2000; div. 2008)​
Talulah Riley
​
​(m. 2010; div. 2012)​
​
​(m. 2013; div. 2016)
'''
```

```python
age_pattern = r'age (\d+)'
age = re.findall(age_pattern,text)
age
```

```python
name_pattern = r'Born *(.*)\n'
name = re.findall(name_pattern,text)
name[0].strip()
```

```python
birthdate_pattern = r'Born.*\n(.*)\(age'
birthdate = re.findall(birthdate_pattern,text)
birthdate[0].strip()
```

```python
birth_place_pattern = r'\(age.*\n(.*)'
birthplace = re.findall(birth_place_pattern,text)
birthplace[0]
```

```python
def get_pattern_match(pattern,text):
  matches = re.findall(pattern,text)
  if matches:
    return matches[0]
```

```python
def extract_personal_information(text):
    age = get_pattern_match('age (\d+)', text)
    full_name = get_pattern_match('Born(.*)\n', text)
    birth_date = get_pattern_match('Born.*\n(.*)\(age', text)
    birth_place = get_pattern_match('\(age.*\n(.*)', text)
    return {
        'age': int(age),
        'name': full_name.strip(),
        'birth_date': birth_date.strip(),
        'birth_place': birth_place.strip()
    }
```

```python
extract_personal_information(text)
```
